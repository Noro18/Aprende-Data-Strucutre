# Array 


## Oinsa atu kria array 

````python

valores = [3, 5, 6, 7] # array ho tipo dados int
naran = ["Antonio", "Jose", "Maria", "Edmilson"] # array ho tipo dados String

print("Array Integer: ", valores)
print("Array String: ",naran)
````

# Operassaun ba iha array

operasaun ba array mak hanesan:

1. Asesu Elementu iha array
2. modifica elemntu iha array
3. Hamos elementu
4. Hatama elementu ka valor foun
5. Buka array nia naruk
6. Halo iterasaun ba iha array

### Asesu elementu iha husi array


````python
# Asesu ho index

print(valores[2])
print(valores[3])
print(valores[1])

print(naran[0])
print(naran[3])

# Asesu range husi list 

print("husi 0 ate 3", valores[0:3])
print("husi 0 ate 3", naran[0:3])
````

### Modifika elementu ida nia valor

maneira atu muda elementu ida nia valor mak 

1. asesu nia elementu
2. attribui fali ho valor seluk

````python
print("valor antes: ", valores)

valores[1] = 10

print("valor depois", valores)

valores[3] = 18

print(valores)
````

### Fo sai Array nia naruk

ita atu fo sai array ida nia naruk iha python, ita uza metodo `len()`

Len husi array sura komesa husi **1** laos hanesan ho index


````python
# Lenght of an array

array = [1, 2, 3, 4, 5, 6]

print(len(array)) # Output: 6
````

### Aumenta elementu foun ba array

### Iterasaun Array

Iterasuan mak ita halo repetisaun procesu ida.

baibain ita uza looping 
- `for`
- `while`




````python
# print manual

print(valores[0])
print(valores[1])
print(valores[2])
print(valores[3])

# print iterative

print("print automatic")
for i in valores:
    print("valor original: ", i)
    print("double: ", i * 2)
    print("")
````

### Exercicio 1

Halo iterasaun ba array ida no print kada elementu ne'e nia par ou impar.

ex:

array = [2, 3, 4, 5]

output =

```
2 numeru par
3 numeru impar
4 numeru par
5 numeru impar
```


````python
array = [2, 3, 4, 5]

for i in array:
    if i % 2 == 0:
        print(i, "numeru par")
    else:
        print(i, "numeru impar")
````
