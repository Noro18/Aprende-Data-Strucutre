# Stack

- Linear Data Structure 
- Rai dados ho maneira LIFO (Last In First Out)
- Katak dados ikus ne’ebe ita hatama nia mak sei sai uluk
- Operasaun basika  husi Stack nia ita hanaran
    - Push (hatama) `append()` iha python
    - Pop (hasai) `pop()` iha python

## Stack Creation


```python
myStack = [] # kria stack ida ho naran myStack


print("Stack original = ", myStack)

myStack.append(3)
myStack.append(5)
myStack.append(7)

print("Stack depios ne'e ita push elemntu foun = ", myStack)

myStack.pop()
myStack.pop()

print("stack depois de pop dala 2 = =", myStack)
```

# Implementasaun Stack iha python

Iha pyhton iha implementasaun 3 ba stack nian 

1. `array`
2. `collections.dequeue`
3. `Queue.lifoqueue`

## Array implementation

```python

# Kria lista foun 
stack = []

stack.append('a')
stack.append('b')
stack.append('x')
stack.append('s')

print("Initial Stack = ", stack)

# pop() elementu husi stack

print(stack.pop())
print(stack.pop())
print(stack.pop())

print('Stack hafoin elementu sira hasai')
print(stack)
```

## `collections.dequeue` Implementation

```python
from collections import deque

stack = deque()

stack.append('a')
stack.append('b')
stack.append('c')
stack.append('d')
stack.append('f')

print("Stack depois de append = ", stack)

print("\nElementu sira ne'eb sei pop sai")
print(stack.pop())
print(stack.pop())
print(stack.pop())

print("\n stack depois de pop dala 3")
print(stack)
```
