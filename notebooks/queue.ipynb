# Implementasaun uza Python List 

Python List bele simula behavious ka karakteristika husi data structure queue nian liu hsui metodo 2

- .`append()`: uza ba enqueue
- `.pop()`: uza ba dequeue

````python
queue = []


# Enqueue
queue.append(10)
queue.append(20)
queue.append(30)

print(queue)
````

````python
# Dequeue

print(queue.pop(0))
print(queue)
````

# `collections.dequeue` implementation

Jadi atu aresolve problem inficiensia ho python list mak ita bele uza `collections.deque`

## saida mak deque? 

`deque` ou **double-ended-queue** mak estrutura dados special ne'ebe permite insersaun no hasai husi oin no kotuk 

## deque la hanesan ho dequeue

**deqeue** ne'e estrutura dados ida no
**Dequeue** ne asaun ida atu hasai elementu husi oin husi fila ka queue ida 



````python
from collections import deque

queue = deque()

queue.append(10)
queue.append(20)
queue.append(30)
queue.append(40)

print(queue)

````

````python
print(queue.popleft())
print(queue.popleft())

print(queue)
````
